```python
#Importing Libraries
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
#Reading Image
image = mpimg.imread('woman.jpg')
plt.imshow(image)
```

#### Converting to Grey Image

```python
#This step is done to reduce complexity of the image
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image, cmap='gray')
```

#### Conversion to  Inverted Grey Image

```python
inverted_image = 255 - gray_image
plt.imshow(inverted_image, cmap='Greys_r')
```

#### Conversion to Blurred Image

```python
# This step is done to smoothen edges
inverted_blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))
ax1.set_title('inverted gray')
ax1.imshow(inverted_image, cmap='Greys_r')
ax2.set_title('inverted blurred image')
ax2.imshow(inverted_blurred, cmap='Greys_r')
```

#### Conversion to Pencil Sketch

```python
final = cv2.divide(gray_image, 255 - inverted_blurred, scale=240)
plt.figure(figsize=(15,15))
plt.imshow(final,cmap="gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.show()
```

#### Showing all images in one frame

```python
plt.figure(figsize=(30,30))
plt.subplot(1,5,1)
plt.imshow(image)
plt.axis("off")
plt.title("Original Image")
plt.subplot(1,5,2)
plt.imshow(gray_image, cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.subplot(1,5,3)
plt.imshow(inverted_image,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.subplot(1,5,4)
plt.imshow(inverted_blurred,cmap="gray")
plt.axis("off")
plt.title("Smoothen Image")
plt.subplot(1,5,5)
plt.imshow(final,cmap="gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.show()
```
